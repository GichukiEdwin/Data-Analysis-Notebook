# The school math in Python
### An arithmetic operator is a mathematical function that takes 2 operands and performs a calculation on them. They include,

  * `+ Addition`: Adds values on either side of the operand
  * `- Subtraction`: Subtracts the right-hand operand from the left-hand operand
  * `* Multiplication`: Multiplies values on either side of the operator
  * `/ Division`: Divides the left-hand operand by the right-hand operand
  * `% Modulo`: Divides the left-hand operand by the right-hand operand and returns the remainder
  * `** Exponent`: Performs exponential (power) calculation on the operands
  * `// Truncation or Floor division`: Returns the division of the operands where the result is the quotient

### Comparison operator - They are used for comparing values. It returns either `True` or `False` according to the condition. They include,
  *  `> Greater than` -  True if the left operand is greater than the right operand
  *  `< Less than` - True if the left operand is less than the right operand
  *  `== Equal to` - True if the left operand is equal to the right operand
  *  `!= Not equal to` - True if operands are not equal
  *  `>= Greater than or equal to` - True if the left operand is greater or equal to the right
  *  `<= Less than or equal to` - True if the left operand is less than or equal to the right

### Logical Operators - This is a symbol or word used to connect 2 or more expressions such that the value of the compound expression produced depends only on that of the original expressions and on the meaning of the operator. They include,
  * `and`: Returns True if both the operands are true
  * `or`: Returns True if either of the operands is true
  * `not`: Returns True if the operand is false

### Membership and Identity Operators
- Membership operators -They are used to test whether a value or variable is found in a sequence(string, List, Tupple, etc) They include,
  * `in`: Returns True if the value is found in the sequence
  * `not in`: Returns True if the value is not found in the sequence
- Identity operators - When a value is stored, it is given a name(variable), which identifies a memory location where the value is stored.
- The `id()` method returns the memory address of the variable.
- The 2 identity operators provided by python check if the 2 values are located on the same part of the memory. They include,
  * `is` -returns True if the operands are identical(refer to the same object)
  * `not is` - returns True if the operands are not identical (do not refer to the same object)

### Walrus operator
- The walrus operator also called the *the new assignment expression* in python, is a new way to assign to variables within an expression using the notation,
```
name:=expr
Colon and equal sign used together.
```


````python
a=5
b=3
c=a+b
print("Addition is", c)

age=16
required_age=18
print(age>required_age)

# logical operators
x=True
y=False
print("x and y is", x and y)
print("x or y is", x or y)
print("not x is", not x)

# membership operators
language="Python"
print("x" in language)
print("z" in language)
print("x" not in language)

# identity operator
num1=5
num2=5
num3=10
print(id(num1))
print(id(num2))
print(id(num3))
print(num1 is num2)
print(num1 is not num2)
print(num1 is num3)
print(num1 is not num3)

# walrus operator 
# convectional way
name="John"
print(name)
# walrus way
name="John"
print(name:="Jake")
````

# Decision Making

### To make decisions, we use a concept called *Conditionals*.
- In python, the conditional statements are,
  * If statement
  * If-else statement
  * If-elif-else statement
  * Nested If-else statement

#### If-else statement
- Used to  check whether a particular condition is True or False. If the condition evaluates to True, then **ONLY** the statements under the if statements will execute, otherwise they will be simply ignored, and statements outside the if block are executed.
```
 if condition:
    # Statements to be executed if the condition evaluates to True
```

#### If-else statement
- An else statement is executed if the condition (boolean expression) is tested by the if block evaluates to false.
```
if condition:
    # Statements to be executed if the condition evaluates to true
else:
    # Statements to be executed if the condition evaluates to false
```

#### If-elif-else statement
- The `if-elif-else` statement is used for handling multiple conditions ie, when ther is more than one condition to check.
- The `elif` statement allows you to check multiple expressions for True and execute a block of code as soon as one of the conditions evaluates to True
```
if condition:
    # statements
elif condition2:
    # statements
elif condition3:
    # statements
else:
    # statements
```

#### Nested If-else
- Nested if construct is used in a situation where you want to check for another condition after a condition resolves to True.
- You can have an if...elif...else construct inside another if...elif...else construct.
- Any number of these statements can be nested inside one another. Indetation is the only way to figure out the level of nesting.

````python
marks=int(input("Enter your marks: "))
if marks>=90:
    print("Grade - Excellent")
elif marks<90 and marks>=75:
    print("Grade - First class")
elif marks<75 and marks >=40:
    print("Grade - Average")
else:
    print("Grade - Fail")

# nested if-else
username=input("Enter your username: ")
password=input("Enter your password: ")
if username=="Gichuki":
    if password=="12345":
        print("Login successful")
    else:
        print("Incorrect password")
else:
    print("Incorrect username")

# traffic light example
light=input("Which color is the light indicating?: ")
if light=="green":
    print("Please keep going")
elif light=="yellow":
    print("Please go slow")
else:
    print("Please stop!")

# game example
user1=8
user2=5
user3=2
if(user1>user2) and (user1>user3):
    print("User one scored the highest points")
elif(user2>user1) and (user2>user3):
    print("User Two scored the highest points")
else:
    print("User Three scored the highest points")

# check if a number is even or odd
num=int(input("Enter a number: "))
if num%2==0:
    print("The number is even")
else:
    print("The number is odd")
````

# Operation on strings

#### Strings are the sequence of characters that can be stored in a variable enclosed in double or single quotes
## String Formatting
- Formatting is important in instances where we want to use the values of variables is our strings to print something or to generate a customized string.
- The following are a few ways to format strings,
  1. *Using comma* - this is generally used when we want to format a string and use the value of the variable at the end. Python adds a space between the words when we use `comma` notation
        ```
        username=input("Please enter your name: ")
        print("Hello", username)
        ```
   2. *Using `str.format`* -Python 3 provides the new `format()` method that does simple positional formatting. This method formats the specified  values and inserts them inside the string's placeholders. The placeholder is defined using the curly brackets: {}
       **Note**: This method requires that the user to take care of spaces in the text. Python will exactly use the string given as the input and replace the placeholders with values
       - The placeholders can be identified using, named indexes `{prices}`, numbered indexes `{0}`, or even empty placeholders `{}`
       ```
       string_one="My name is {lname}, I am {age} years old". format(lname="Gichuki", age=23)
       string_two="My name is {0}, I am {1} years old". format("Gichuki", 23)
       string_three="My name is {}, I am {} years old". format("Gichuki", 23)
       ```
    3. *F-strings* - f-string is a literal string, prefixed with 'f', which contains expressions inside braces, the expressions are replaced with their values
       - they provide a way to embed expressions inside string literals, using a minimal syntax
       - the string itself can be formatted in much the same way that you would with `str.format()`
       - the f-string is really an expression evaluated at runtime, not a constant values
            ```
            name=input("Enter your name:")
            print(f"Hello, {name}!")
            ```
       - this new formatting syntax is quite powerful
       - you van even embed arbitrary expressions in the fstrings and perform various  kinds of operations
         ```
         num1=5
         num2=10
         print(f"Five plus ten is {num1+num2}")
         ```

### Concept of Index
- A string is a sequence of characters, and an index is a numerical representation of an item's position in a sequence
- The sequence can refer to many things: a list, a string of characters, or any arbitrary sequence of values
- The numerical representation starts the count from 0 in Python

#### Accessing values
- If you need to access a particular value from a string, pass the index number of the required element in square brackets
*syntax:-*
      ```
      variable_name[index_number]
      ```


#### Slicing
- Slicing is extracting a part of the string from the specified index
- just like accessing anelement, we can slice a string by using the index number
*syntax:-*
```
string[start:end:step]
```
- slicing starts with the start index (included) and ends at the end index (excluded)
- the step parametr is used to specify the steps to take from the start to the end index and is optional
- if we do not pass any of the start or end values, it considers the whole string from the start to the end index respectively
- when we pass the start and end value, it starts from the start index and stops by excluding the end index..

# Looping

- Looping is an important concept when we need to execute a bunch of statements based on some conditions repeatedly
- Statements are executed sequentially
- Meaning the statement in a bloc is executed first, followed by the second, and so on.
- There can be a situation where we need to execute a block of code several number of times
- A loop statement allows us to execute a statement or a group of statements multiple times
- **Repetitive** execution of the same block of code over and over is reffered to as iteration
- There are two types of iteration,
    1. **Definite iteration**, in which the number of repititions is specified explicitly in advance
    2. **Indefinite iteration**, in which the code block executes until some condition is met
- Python provides various loops to handle theiterations, namely,
    1. **For loop** - it is generally used for definite iteration
    2. **While loop** - it is generally used for indefinite iteration

## **range()**
- The range() is a built-in function that is used when a user needs to perform an action for a specific number of times
- It is used to generate a sequence of numbers and is commonly used with the for loop ie. `range` allows the user to generate a seriesof numbers within a given range
- Depending on how many arguments are passed to the range function, we can decide where the series of numbers will begin and end as well as what the step size will be between a number and the next
- The function mainly takes 3 arguments:
    1. **start**: an integer starting from which the series of integers is to be returned
    2. **stop**: an integer before which the series of integers is to be returned, the range of integers ends at stop -1
    3. **step (optional)**: an integer value that determines the increment between each integer in the sequence

## **for loop**
- for loop is used for sequential traversal
- it has ability to iterate over the items of any sequence, such as a list or string
    ```
    for val in sequence:
        statements
    ```
- val is the variable that takes the value of the itam inside the sequence on each iteration
- loop continues until it reaches the last item in the sequence, the body of for loop is separated from the rest of the code using indentation
*example one*
    ```
    for i in range(5)
        print(i)
    ```
- *i is the iterating variable* - it iterates in the given range and executes the statements in the body of the loop
- it will start from 0 and run till range-1 because the stop value in a range function is excluded

## while loop
- It is used to execute a block of statements repeatedly until a given condition is met
- if the condition becomes false, the line immediately after the loop in the program will be executed, this is the syntax
    ```
    while boolean_expression
        statement(s)
    ```

- boolean_expression is first evaluated  and if it is true, only then the loop body is executed
- next, the expression is evaluated again, if it is true, then the body again and this continues until the expression becomes false


## Break statement
- The break statement is used to terminate the loop statement in which it is present, the control then passes to the staements that are present after the loop block ends
- The break statement is important where a condition arises and you want to exit the loop completely before the end condition arrives, it changes the execution from its normal sequence
- If it is present in the nested loop, then it terminates only those loops which contain a break statement
    ```
    str="Goodmorning"
    for i in str:
        if i==="m":
            break
        else:
            print(i)
    ```

- the program iterates over a string "Goodmorning", when we iterate, we just need to print "Good" and skip the iteration after it
- thus every time we need to check the value if "i" and when the value of "i" is equal to "m", we need to stop the execution of the loop and come out of it

````python

#? range()
print(list(range(5))) # single argument signifies the end of the range(excluded)
print(list(range(1,5))) # start and end
print(list(range(1,10,3))) # start, end and step size

#! for loop
for i in range(3):
    print("Gichuki")

#? iterating over a string
for i in "Gichuki":
    print(i)

#! while loop
i=0
while i<3:
    print("Gichuki")
    i+=1

#? break statement
boarding_capacity=int(input("Enter the boarding capacity: "))
threshold=boarding_capacity*(50.0/100.0)
for i in range(0,boarding_capacity,10):
    if i>threshold:
        print('50% capacity reached')
        break
    else:
        print("Passenger onboarded:",i)
````

# List
- Lists in Python is a type of variable which allows us to store multiple values of various data types in a single variable
- It is a sequence type data structure just like a string
- We can access the values individually using the concept of index numbers
- Important features of a List are:-
    * Lists are ordered collections of multiple values
    * Lists are mutable, ie, values of lists can be changed and manipulated over a period of time
    * Lists allows us to store duplicate values
    * Lists are created by placing the sequence inside the square brackets[...]
- The elements inside the sequence are comma separated
- It can have any number of items and they may be of different types (integer, float, string, etc)
    ```
    list1=[]
    list2=[1,4,5,7,3,9]
    list3=["Edwin",23,34.9,True]
    ```

### Manipulating a list
1. Accessing elements in a list
2. List slicing
3. Updating a list
4. Deleting elements from a list
- Since the list is a sequence data type, we can access elements of a list using indices
- Just like strings, the elements of a list are assigned with an index starting from zero

### Operations on lists
- concatenation
- repitition
- membership

- Adding two lists is list concatenation and multiplying a list is repitition
- We use `'+'` and `'*'` operators respectively
    ```
    lst1=[1,2,3]
    lst2=[4,5,6]
    print(lst1+lst2)
    print(lst1*2)
    ```

- Membership operator `(in)` is used to test the membership of an element in a list

### built-in functions for lists
* len(list)- returns the length of the list
* min(list)- returns the elements of the list with the minimum value
* max(list)- returns the elements of the list with the maximum value

### lists methods
- methods can be accessed using the syntax `list.method()`, they include,
    1. append(element): adds the specified element to the end of the list
    2. insert(element): inserts the element at the specified position or index
    3. pop(): removes and returns the last element from the list
    4. remove(element): removes the specified element from the list
    5. reverse(): reverse the order of elements in the list
    6. index(element): returns the index of the first matched item
    7. count(element): returns count of how many times an element occur in the list

### smart lists - list comprehension in python
- There are istances where we want to create lists at runtime or create a list from another list or to create a list depending on user data
- Let's for example assume we want to create a list of even numbers at runtime, this is how to do it
    ```
    even=[]
    for x in range(1,11):
        if x%2==0:
            even.append(x)
    print(even)
    ```

- List comprehension is an elegant way to define and create a list in Python
- We can create lists just like mathematical statements and in only a single line
- The syntax for list comprehension is easier to grasp, it consists of the following parts:
    1. output expression
    2. input sequence
    3. a variable representing a member of the input sequence
    4. an optional predicate part
- using list comprehension to create a list of even numbers in a given range,
    ```
    even=[x for x in range(1,11) if x % 2 ==0]
    print(even)
    ```

- Here,
    * > `x` is the output expression
    * > `range(1,11)` is the input sequence
    * > `x` is variable 
    * > if `x % 2 == 0` is predicate part/condition

- ***Note***: Not all parts of a list comprehension are mandatoryto be used, look 👇👇
    ```
    string="Hello, list"
    characters_list=[x for x in string]
    print(characters_list)
    ```
- ***Note***: There is no condition mentioned in the list comprehension part

````python

#? accessing elements in a list
names=["Edwin","Gichuki","Eduard","Jenny","Barry"]
print(names)
print(names[0])
print(names[2])

#! slicing
print(names)
print(names[0:3])
print(names[1:4])
print(names[2:])

#! sorting
names.sort()
print(names)
names.sort(key=len)
print(names)

#? update
marks=[56,76,54,89,90]
print(marks)
marks.sort()
print(marks)
marks[4]=100
print(marks)

seq=[7,2,3,7,5,6,0,1]
seq[3:4]=[6,3]
print(seq)

#! delete
names=["Edwin","Gichuki","Eduard","Jenny","Barry"]
print(names)
del names[1]
print(names)

#? iteration on lists
fruits=["apple","banana","cherry","kiwi","mango"]
for x in fruits:
    print(x)

#! checking for a value
fruits=["apple","banana","cherry","kiwi","mango"]
for i in fruits:
    if i=="banana":
        print("banana is in the list")
        break
    else:
        print("banana is not in the list")

#? functions
numbers=[23,34,45,56,67,78,89]
print(min(numbers))
print(max(numbers))
print(sum(numbers))
print(len(numbers))
print(sorted(numbers))
print(sorted(numbers,reverse=True))

#! list comprehension
even=[x for x in range(1,11) if x % 2 ==0]
print(even)

string="Hello, list"
characters_list=[x for x in string]
print(characters_list)
````

# Read-Only Lists: Tuples

### Tuples are similar to the lists. Lists allow us to store multiple values in a single variable
+ The only major difference is that values of lists can be changed, ie, Lists are mutable, whereas Tuples are immutable, ie, values of Tuples cannot be changed
+ Tuples are used in situations where we may wish to restrict the manipulation of sequences of values
+ We can only fetch and read the values
+ Tuples is a data type/structure in Python, they come under the category of sequence data types
+ Here are some important features of tuples
    * Tuples are ordered collections of multiple values
    * Tuples are immutable
    * Tuples allow us to store duplicate values

#### creating a tuple
+ Tuples are created by just placing the sequence inside the `paranthesis ()`
+ The elements inside the parentheses are comma separated
+ it can have any number of items and they may be of different types (integer, float, string, etc...)
+ Example
    ```
    tuple1=()
    tuple2=(1,4,6,8,9)
    tuple3=("Gichuki", 23.5, 56, False)
    ```

#### manipulating tuples
+ the following are ways to manipulate tuples
    1. Accessing elements of a tuple
    2. Tuple slicing
    3. Updating a tuple (Spoiler alert: this is not possible)
    4. Deleting elements of a tuple

+ Accessing elements and tuple slicing - tuples are a sequence data type, thus each element is associated with an index that starts from zero.
+ Just like lists, we can use indexes to access and slice elements of a tuple.
+ Updating tuples - tuples are immutable, thus we cannot update tuples.
+ the code below will raise an error since tuples are immutable:
    ```
    tuple1 = (1, 2, 3, 5, 6, 7)
    tuple1[3] = 9
    print(tuple1)
    ```
**Output**
    ```
    TypeError: 'tuple' object does not support item assignment
    ```

+ Since tuples are immutable, deleting/removing individual tuple elements isn't possible
+ However, you can delete the entire tuple using the `del` keyword

#### operations on tuples
+ All the basinc operations performed on a list can be performed on tuples as well
+ However, the result of the operation is always a new tuple since we cannot manipulate the existing tuple
+ operations,
    1. concatenation
    2. repitition
    3. membership

#### built-in functions for tuples
* len(tuple) - returns the length of the tuples
* min(tuple) - returns the elements of the tuples with the minimum value
* max(tuple) - returns the elements of the tuples with the maximum value
+ since tuples are immutable, we cannot use methods such as, `append(), insert(), remove(), pop(), etc` which manipulate and modify the sequence

+ There are 2 methods that we can use on tuples
    1. `index(element)`: returns the index of the first matched item
    2. `count(element)`: returns the number of times an element occurs in the sequence


#### advantages of tuples
1. we generally use tuples for heterogeneous (different) data types and lists for homogeneous (similar) data types
2. since tuples are immutable, iterating over them is faster than iterating over a list, so there is a slight performance boost
3. tuples that contain immutable elements can be used as a key for a dictionary, with lists, that is not possible
4. if you have data that does not change, implementing it as a tuple will guarantee that it remains write-protected


````python


#? accessing and slicing elements of tuples
languages=("Python", "Java", "JavaScript", "Ruby", "Rust")
print(languages)
print(languages[3])

#! slicing
print(languages[1:4])

#? concatenation and repitition
tuple1=(1,2,3)
tuple2=(4,5,6)
print(tuple1+tuple2)
print(tuple1*2)

#! are you a member
print("Rust" in languages)
print("Go" in languages)

#? iterating over a tuple
for language in languages:
    print(language)

#? built-in functions and methods
numbers=(78,24,56,78,29,49,78)
print(len(numbers))
print(min(numbers))
print(max(numbers))
print(numbers.count(78))
print(numbers.index(29))
````

# dictionary key-value pairs
+ In programming, there are situations where we need to store data in the form of key-value pairs
+ keys are associated with some values just like in the dictionary which has data in key-value pairs(keys are the words, and the values are the meaning of the words)
+ searching for a particular value  becomes easy with the help of a key
+ similarly, in python we can do that with the help of a data structure called dictionary

#### features of the dictionary
1. dictionary is an unordered and mutable collection of items
2. each item of a dictionary has a key/value pair
3. one can retrieve values when the key is known
4. each in the dictionary is unique, while values can be repeated
#### structure of a dictionary
1. stores values in the form of key-value pairs
2. values of a dictionary can be of any type, but the keys must be of an immutable data type such as strings, numbers, or tuples
#### creating a dict
+ each key is separated by a colon, the items are separated by commas, and the whole thing is enclosed in curly braces
```
# empty dict
dict1={}
# dict with int keys and string values
dict2={1:"Edwin",2:"Sam"}
```

+ dictonaries are unordered and order is not important
+ keys play an important role while manipulating and accessing dictionaries

### manipulating dictionaries
+ dictionaries are unordered, and thus we cannot use the concept of an index to access and manipulate elements in a dict
+ instead we use the concept of keys since those are unique, and so the values can be uniquely identified with the help of their keys
+ dictionaries can be manipulated by performing the following,
  1. accessing elements of a dict
  2. updating a dict
  3. deleting elements from a dict
1. accessing elements
   + keys can be used either inside `square brackets []` or with the `get()` method
   + if we use the `square brackets []`, `KeyError` is raised in case a key is not found in the dict
   + on the other hand, the `get()` method returns `None` if the key is not found
2. updating dictionaries
    - dictionaries are mutable, hence we can manipulate their values
    - we can either update the value or add a new key/value pair to the collection
    - if the key is already present, then the existing value gets updated, in case the key is not present, a new (key: value) pair is added to the dict
3. deleting elements
    + just like lists and tuples, we can use the `del` keyword to delete a particular item or delete the whole dict
    + we can use `del` to delete the entire dict

### dictionary functions
+ python provides functions to manipulate the dict, a few are,
    1. `key()`: returns the list of dictionary keys
    2. `values()`: returns the list of dictionary values
    3. `clear()`: removes all elements of a dictionary, that is, it makes an empty dict
    4. `get(key)`: returns the value of the specified key
    5. `item()`: returns a list of dictionaries key-value pairs in tuple form
    6. `update(dict)`: adds specified dict(items) key-values pairs to the dictionary
#### updating list-based values
+ assume we have a key with 10 different values stored in a list format, if we need to either add or remove a particular value in th list, we use,
    * `append()`
    * `remove()`

````python

#? dict - elearning website
students_enrolled={"John":"Python", "Sam":"Rust","Nick":"Go","Willis":["Python","JavaScript","C"]}
print(students_enrolled)

#? accessing elements
print(students_enrolled["John"])
print(students_enrolled.get("Willis"))

#? update
students_enrolled["John"]="Scala"
students_enrolled["Faith"]="Java"
print(students_enrolled)

#? delete
#! del students_enrolled["Faith"]
#? print(students_enrolled)

#? dictionary methods
print(students_enrolled.keys())
print(students_enrolled.values())
print(students_enrolled.items())

#? updating list-based values
students_enrolled["Willis"].append("C++")
# students_enrolled["Willis"].remove("JavaScript")
print(students_enrolled)

#? iteration
for i in students_enrolled:
    print(i)

for i in students_enrolled.values():
    print(i)

# #? accessing both key and value while iterating
for i in students_enrolled.items():
    print(i)
````

# sets
+ set is a collection - grouping objects into a set can be useful
+ python provides a built-in set type to do so
+ just like tuples, lists, dictionaries, `Set` is a data structure in python that can be used to group and store multiple elements

#### features of sets
- sets are unordered collections, thus there is no index attached to any element in a python set
- set elements are unique, duplicate elements are not allowed
- a set can be modified, but elements contained in the set must be of an immutable type

+ a set is created using the `set()` function or placing all the elements within a pair of curly braces

#### accessing data
+ sets are unordered, hence we cannot use index numbers to access elements based on index
+ thus, we can not access an individual element, but we can access elements one by one by looping through it

+ set specific operations include,
    + union of sets - the union of 2 sets A and B is the ste of elements that are in A, in B or both A and B
    + set intersection - the intersection of 2 sets A and B is the set containing all elements of A that also belongs to B, ie, the intersection of sets A and B is a set of elements that are common in both sets
    + difference of sets - the difference operation on 2 sets produces a nwe set containing only elements from the first set and none from the second set


````python

#? creating a set
set1=set() # empty set
set2={"Python", "Java", "JavaScript", "Ruby", "Rust"}

#? accessing elements
print(set2)
print("Python" in set2)

days={"Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"}
print(days)

for day in days:
    print(day)

#? set union
a={1,2,3,4,5}
b={4,5,6,7,8}
print(a.union(b))
print(a|b)

#? set intersection
print(a.intersection(b))
print(a&b)

#? set difference
print(a.difference(b))
print(a-b)
````
