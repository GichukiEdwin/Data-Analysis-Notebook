#### descriptive statistics of the dataset
+ sums, average, medians, and modes of the population and death analysis

```python
# summary statistics
df[['pop_est_2018','pop_est_2019','births_2019','deaths_2019']].describe()
```

```python
# Calculate sums
sums = df[['pop_est_2018', 'pop_est_2019', 'births_2019', 'deaths_2019']].sum()

# Calculate modes
modes = df[['pop_est_2018', 'pop_est_2019', 'births_2019', 'deaths_2019']].mode()

print("Sums:\n", sums)
print("Modes:\n", modes)
```

#### alternatively

```python
# Calculate sums
print(f"Total Population 2018: {df['pop_est_2018'].sum()}")
print(f"Total Population 2019: {df['pop_est_2019'].sum()}")
print(f"Total Deaths 2019: {df['deaths_2019'].sum()}")
print(f"Total Births 2019: {df['births_2019'].sum()}")

# Calculate averages
print(f"Average Population 2018: {df['pop_est_2018'].mean()}")
print(f"Average Population 2019: {df['pop_est_2019'].mean()}")
print(f"Average Deaths 2019: {df['deaths_2019'].mean()}")
print(f"Average Births 2019: {df['births_2019'].mean()}")

# Calculate medians
print(f"Median Population 2018: {df['pop_est_2018'].median()}")
print(f"Median Population 2019: {df['pop_est_2019'].median()}")
print(f"Median Deaths 2019: {df['deaths_2019'].median()}")
print(f"Median Births 2019: {df['births_2019'].median()}")

# Calculate modes
print(f"Mode Population 2018: {df['pop_est_2018'].mode()[0]}")
print(f"Mode Population 2019: {df['pop_est_2019'].mode()[0]}")
print(f"Mode Deaths 2019: {df['deaths_2019'].mode()[0]}")
print(f"Mode Births 2019: {df['births_2019'].mode()[0]}")
```
